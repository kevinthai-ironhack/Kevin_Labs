We break down the main problem into several sub problems:

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
# Sub Problem 1: Aggregate the amount_spent for unique customers

#orders_agg = orders.groupby('CustomerID')[['amount_spent']].agg('sum')
orders_agg = orders.groupby('CustomerID')[['amount_spent']].sum()
orders_agg.sort_values(by='amount_spent', ascending=False).head()
```

```python
# Sub Problem 2: How to select customers whose aggregated amount_spent is in a given quantile range?
# bins = pd.cut(students_df['py-score'],4,labels=score_labels,retbins=True)
orders_agg.describe()
```

```python
# criteria for VIP
np.quantile(orders_agg,0.95)
```

```python
# criteria for Preferred 
np.quantile(orders_agg,0.75)
```

```python
score_labels=["Normal","Preferred","VIP"]
bins = pd.DataFrame(pd.cut(orders_agg['amount_spent'],[orders_agg['amount_spent'].min(),np.quantile(orders_agg,0.75),np.quantile(orders_agg,0.95),orders_agg['amount_spent'].max()],labels=score_labels))
#bins = pd.cut(students_df['py-score'],[70,80,90,100,108], labels=score_labels,retbins=True)
bins
```

```python
cust_bins = pd.merge(orders_agg, bins, left_index=True, right_index=True)
cust_bins.rename(columns={'amount_spent_y':'classification'},inplace=True)
cust_bins.sort_values(by='amount_spent_x', ascending=True).head()
```

```python
cust_bins['classification'].value_counts()
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
cust_bins
```

```python
#df = df.drop_duplicates(subset=['C1','C2','C3'])
country = orders[["CustomerID","Country"]]
country.drop_duplicates(subset=['CustomerID',"Country"],inplace=True)
country.set_index("CustomerID",inplace=True)
country
```

```python
cust_bins_country = cust_bins.merge(country, on='CustomerID', how='left')
# left_df.merge(right_df, on='user_id', how='left')
cust_bins_country
```

```python
df_vip = cust_bins_country[cust_bins_country["classification"]=="VIP"]
df_vip
```

```python
df_vip.reset_index().groupby(['Country'])[['CustomerID']].count().sort_values(by="CustomerID",ascending=False).head(5)
#result = df.groupby(['Courses','Fee'])['Courses'].count()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
df_vip_pref = cust_bins_country[(cust_bins_country["classification"]=="VIP")|(cust_bins_country["classification"]=="Preferred")]
df_vip_pref
```

```python
df_vip_pref.reset_index().groupby(['Country'])[['CustomerID']].count().sort_values(by="CustomerID",ascending=False).head(5)
```
