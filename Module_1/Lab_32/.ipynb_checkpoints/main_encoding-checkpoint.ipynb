```python
import pandas as pd
import numpy as np
# 
d = {
    'TransactionID':[1,2,3,4,5],
    'ClientID':[231,765,453,231,892],
    'Profession':['Self-employed','students','Horeca','Self-employed','finance'],
    'Bank_dep':['009','005','007','009','003'],
    'Risk':['high','high','medium','high','low'],
    'Number of credits':[2,3,5,2,3],
    'Revenue':[30200,12700,89400,30200,740000]
}
df = pd.DataFrame(data=d)
```

```python
df
```

```python
# encoding categorical data which are: profession, bank_dep, risk and number of credits ?
```

### ONE HOT ENCODING

```python
# get dummies
# pd.get_dummies(df, prefix=['col1', 'col2'])

pd.get_dummies(df,columns=['Profession','Bank_dep','Risk','Number of credits'])
```

### LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder

#create instance of label encoder
lab = LabelEncoder()

#filtering df with NOMINAL columns only
df_label = df[['Profession','Bank_dep']]

#perform label encoding on categorical data:
df_label['Profession_filtered'] = lab.fit_transform(df_label['Profession'])
df_label['Bank_dep_filtered'] = lab.fit_transform(df_label['Bank_dep'])
```

```python
df_label
```

### ORDINAL ENCODING

```python
#filtering df with ORDINAL columns only
df_ordinal = df[['Risk','Number of credits']]
df_ordinal
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enc = OrdinalEncoder()
enc.fit(df_ordinal[['Risk','Number of credits']])
df_ordinal[['Risk','Number of credits']] = enc.transform(df_ordinal[['Risk','Number of credits']])
df_ordinal
```
