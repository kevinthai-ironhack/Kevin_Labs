### 2. Use crosstabs to show the number of moovies per country

```python
# For such kind of count table I prefer to use groupby 
movie.groupby(['country'])[['movie_title']].count().sort_values(by="movie_title",ascending=False).head(10)
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
# index = director, columns = country, values = number of films
```

```python
pd.crosstab(movie.director_name, movie.country, values=movie.movie_title, aggfunc='count',margins=True, margins_name="Total").fillna(0)
```

### 4. Average duration of filmes grouped by color

```python
# If there is no several columns I prefer to go this way
movie.groupby(['color'])[['duration']].mean().round(0)
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
movie.groupby(['director_name'])[['cast_total_facebook_likes']].mean().round(0).sort_values(by="cast_total_facebook_likes",ascending=False).head(10)
```
